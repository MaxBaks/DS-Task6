```python
import pandas as pd
```

```python
#1
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def getRegion(city):    
    result = 'undefined'
    for q,w in geo_data.items():
        for t in w:
            if t in city['keyword']:
                result = q
                break

    return result
```

```python
keywords = pd.read_csv('keywords.csv')
keywords['region'] = keywords.apply(getRegion, axis = 1)
keywords.head(10)
```

```python
#2
data = pd.read_csv('ratings.csv')
data.head()
```

```python
def classifyRating (rate):
    if rate['rating'] <= 2:
        rateClass = 'Низкий'
    elif 2<rate['rating']<=4:
        rateClass = 'Средний'
    else:
        rateClass = 'Высокий'
    return (rateClass + ' рейтинг')
```

```python
data['class'] = data.apply(classifyRating, axis = 1)
data.head(10)
```

```python
grouped = ratings.groupby('userId').agg(['min','max','count']).reset_index()
grouped.head()
```

```python
grouped.drop(['rating'], axis = 1, inplace = True)
```

```python
grouped.head()
```

```python
ratings_filtered = grouped[grouped['timestamp']['count'] >= 100]
ratings_filtered.head()
```

```python
#4
years = list(range(1950, 2011))
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
movies.head()
```

```python
moviesWithRates = movies.merge(ratings, how = 'left', on = 'movieId')
moviesWithRates.head()
```

```python
def production_year(title):
    for n in years:
        if str(n) in title['title']:
            result = n
            break
        else:
            result = 1900
    return result
```

```python
moviesWithRates['year'] = moviesWithRates.apply(production_year, axis = 1)
moviesWithRates.head(10)
```

```python
grouped1 = moviesWithRates.groupby('year').mean().reset_index()
grouped1.sort_values(by ='rating',ascending = False).head(30)
```
